```python
import pandas as pd
import matplotlib.pyplot as plt
datas = pd.read_csv(r"C:\Users\ms900\OneDrive\Desktop\sales_datas.csv")
print(datas)
```

```python
df =pd.DataFrame(datas)
print(df)
```

 ## Remove Duplicates

```python
df.drop_duplicates(inplace=True)
print(df) 
```

```python
# Sub-category blanks should be filled with “not mentioned” 
```

```python
df["Sub-Category"].fillna("not mentioned", inplace=True)
print(df)  
```

```python
mean= df["Quantity"].mean()
df["Quantity"].fillna(mean, inplace=True)
print(df)
```

```python
# Avg sales 
Avg=df["Sales"].agg({"Avg :":"mean"})
print(Avg)
```

```python
# Mininum Quatity
Min =df["Quantity"].agg({"Min_Qty :":"min"})
print(Min)
```

```python
# Mode of Quantity 
mode = df["Quantity"].agg({"Mode :": "mode"})
print(mode)
```

```python
# Standard deviation of quantity  
std = df["Quantity"].agg({"Standart_diviation :":"std"})
print(std)
```

```python
# Correlation between sales and Profit
Correlation = df["Sales"].corr(df["Profit"])
print("Correlation between sales and Profit : ",Correlation)
```

```python
# Correlation
import matplotlib.pyplot as plt
plt.scatter(df["Sales"],df["Profit"],color="r")
plt.title("Correlation sales and Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
# Create a function – Need to add a new column as “Status” if quantity is greater than 5 then “High sales” else “low sales”
```

```python
def fun(Quantity):
    if Quantity >= 5:
        return "High Sales"
    else:
        return "Low Sales"
df["Status"] = df["Quantity"].apply(fun)  
print(df)
```

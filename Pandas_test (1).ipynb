```python
# Loading Data
```

```python
import pandas as pd
data = pd.read_csv(r"C:\Users\ms900\OneDrive\Desktop\Sales_Data.csv")
print(data)
```

```python
# Data Cleaning - Identify and handle missing values
```

```python
# Quantity missing values should ne filed with mean quantity 
```

```python
mean_value = data["Quantity"].mean()
data["Quantity"] = data["Quantity"].fillna(mean_value)
print(data)
```

```python
# Missing region should be filled with “central” 
```

```python
data["Region"].fillna("central")
print(data)
```

```python
# Duplicates should be removed 
```

```python
data.drop_duplicates()
```

```python
# Missing Sub-category should be filled with “Not mentioned”
```

```python
data["Sub-Category"].fillna("Not mentioned")
```

```python
# Filler the data only sales >1000
```

```python
data2 = data[data["Sales"]>1000]
print(data2)
```

```python
# Sort the data based on sales
```

```python
data3 = data2.sort_values(by = "Sales",ascending=True)
print(data3)
```

```python
data3.to_csv(r"C:\Users\ms900\OneDrive\Desktop\Output.csv",index=False)
```
